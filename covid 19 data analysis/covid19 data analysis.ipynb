### Top 10 total cases

```python
top_10_cases = data.groupby(by = 'Country/Region').max()[['TotalCases', 'TotalDeaths']].sort_values(by =['TotalCases'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 Country with maximum cases")
ax = sns.barplot(data = top_10_cases.iloc[:10], y = "TotalCases", x = "Country/Region", linewidth = 2, edgecolor = "red", palette = ['tab:blue', 'tab:orange', 'tab:red', 'tab:green', 'tab:purple', 'tab:pink']) #iloc= indexlocation
plt.xlabel("Country/Region")
plt.ylabel("Total Cases")
plt.show()
```

```python
top_10_deaths = data.groupby(by = 'Country/Region').max()[['TotalCases', 'TotalDeaths']].sort_values(by =['TotalDeaths'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 total deaths", size = 25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y ="TotalDeaths", x ="Country/Region", linewidth = 2, edgecolor= 'black')

plt.xlabel("Country/Region")
plt.ylabel("TotalDeaths")
plt.show()
```
