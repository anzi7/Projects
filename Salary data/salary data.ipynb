## Convert k into 1000

```python
# Extract numerical values from 'Salary' column and format as 44000-99000
df['Salary'] = df['Salary'].str.replace('k','000')
```

```python
df.head()
```

```python
# Remove '$' from 'Salary' column
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True)
```

```python
df.head(5)
```

```python
df.head(20)
```

```python
# Replace 'NaN' in 'Age' column with the mean of the column
df['Age'] = df['Age'].fillna(df['Age'].mean()).astype(int)
df.head(20)
```

Replace

```python
df['Location'].value_counts()
```

```python
df['Location'] = df['Location'].replace({'New York,Ny':'New York', 'India,In':'India','India In':'India', 'Australia Aus':'Australia'})
df.head(10)
```

```python
df['Location'] = df['Location'].str.replace(r'\sIn$', '')
```

```python
# Replace -1.0 with NaN in 'Established' column
df['Established'] = df['Established'].replace(-1.0, float('0'))
df.head(5)
```

```python
# Replace -1 with 'FALSE' in 'Easy Apply' column, ensuring 'Easy Apply' is treated as string
df['Easy Apply'] = df['Easy Apply'].astype(str).replace('-1', 'FALSE')

# Display the DataFrame
df.head(5)
```

```python
# Replace -1.0 with NaN in 'Rating' column
df['Rating'] = df['Rating'].replace(-1.0, float('nan'))
df.head(10)

# Display the DataFrame
```
